### 1.HealthCare Plus recorded the daily number of patient admissions for the past 10 days:
### [32, 28, 35, 30, 29, 27, 31, 34, 33, 30]

### Compute the mean, median, and mode of patient admissions.
### Which measure best represents patient admissions?
### If the hospital increases its admission capacity by 10%, how will this affect the measures of central tendency?

```python
## Take the given list in form of a numpy array
patient_admits = np.array([32, 28, 35, 30, 29, 27, 31, 34, 33, 30])

print(Fore.GREEN + Style.BRIGHT + "The central tendency(mean,median,mode) calculation")
## Take the mean and print it
mean = np.mean(patient_admits)
print(Fore.BLUE + f"   The average of patient admission is: {mean}")

## Take the median and print it
median = np.median(patient_admits)
print(Fore.BLUE + f"   The median of patient admission is: {median}")

## Take the mode and print it
mode = stats.mode(patient_admits)[0]
print(Fore.BLUE + f"   The mode of patient admission is: {mode}")

## Out of all different measures which one is better depends on what type of data and values we have
print(Fore.GREEN + "\n\nOut of all different measures which one is better depends on what type of data and values we have,below explanation is what each measure do")
print(Fore.BLACK + "  Mean will be affected by outliers, so if there's any extreme value, it may work inappropriately.")
print(Fore.BLACK + "  Median is not affected by outliers and gives a better idea of the middle value.")
print(Fore.BLACK + "  Mode is the most frequent admission but might not always reflect the typical value, especially if no value repeats frequently.")

##Icrease the patient admission by 10%
icreased_admits = [i+ 0.1 * i for i in patient_admits]

print(Fore.GREEN + "\n\nIf the hospital increases its admission capacity by 10%,the impact on central tendency will be")

## Take the mean and print it
mean = np.mean(icreased_admits)
print(Fore.BLUE + f"   The average ofpatient admission after 10% increase is: {mean}")

## Take the median and print it
median = np.median(icreased_admits)
print(Fore.BLUE + f"   The median of patient admission after 10% increase is: {median}")

## Take the mode and print it
mode = stats.mode(icreased_admits)[0]
print(Fore.BLUE + f"   The mode of patient admission after 10% increase is: {mode}")

print(Fore.GREEN + "\n\nFinding is:")
print(Fore.BLUE + "   We could able to see central tendency before and after increase of patient admissions, seems like the central tendency have increased when the admission increases. Also based on different situation we'll choose the measure.")
```

### 5. The hospital claims that the average patient wait time in the emergency department is 30 minutes. A sample of 10 patient wait times (in minutes) is recorded:
### [32, 29, 31, 34, 33, 27, 30, 28, 35, 26]

### Test whether the hospital’s claim is valid at a 5% significance level.
### State the null and alternative hypotheses.
### If the wait time significantly exceeds 30 minutes, what changes should the hospital implement to reduce waiting time?

```python
## State the Null and Alternate Hypothesis
print(Fore.GREEN + Style.BRIGHT + "Step 1: State the Hypotheses:")
print(Fore.BLACK + "Null hypothesis is: The average wait time in emergency department is 30 minutes.")
print(Fore.RED + "                              μ = 30                                 ")

print(Fore.BLACK + "Alternate hypothesis is: The average patient wait time is not equal to 30 minutes.")
print(Fore.RED + "                               μ ≠ 30                                 ")

## Testing the hospital's claim at 5% significance level
print(Fore.GREEN + "\n\nStep 2: Testing the hospital's claim at 5% significance level, using T-test.")
print(Fore.GREEN + "\n                  Step 2.1: Calculate the Sample Mean and Standard Deviation:")
sample = np.array([32, 29, 31, 34, 33, 27, 30, 28, 35, 26])
x = np.mean(sample)
std = np.std(sample)
print(Fore.BLUE + f"                          sample_mean(x) = {x}                                                   ")
print(Fore.BLUE + f"                          sample_standard_deviation(std) = {std}                                                   ")

print(Fore.GREEN + "\n                 Step 2.2: As number of samples is less than 30, so we'll do a t-test.Calculate the T-Statistic:")

t_statistic_formula = "t = (x̄ - μ) / (s / √n)"
print(Fore.MAGENTA + f"                          t-statistic formula: {t_statistic_formula}\n")

n = 10
mu = 30
t_stat = (x - mu) / (std / np.sqrt(n))

print(Fore.BLUE + f"                          T_Statistics = {t_stat}                 ")
print(Fore.BLACK + "\n                 Here, we need to do a 2-tail test.")
print(Fore.BLACK +"\n                 From T-table for two tailed test, need to check score based on degree of freedom(n-1) = 9 here and significance level 0.05")
print(Fore.BLACK +"\n                 T-score we got from T-table as: 2.262")
print(Fore.BLACK +"\n                 As two tail test so, if t_stat lies within range of -2.26 to +2.26 then fine else not")
print(Fore.BLUE +"\n    As, t statistics for sample << T-score from T-table, means 0.55 is within the critical range not outside, so we fail to reject null hypothesis.")

print(Fore.GREEN + "\n\nStep 3: Testing the hospital's claim at 5% significance level, using P-value from T-test.")
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), 9))
print(Fore.BLUE + f"                          p_value = {p_value}                ")
print(Fore.BLUE + f"                          significance_level(α) = 0.05       ")

# Interpretation
print(Fore.GREEN + "\n\nStep 4: So the interpretation is:")
if p_value < 0.05:
    print(Fore.BLUE + "                    The average wait time in emergency department is more than 30 minute, reject null hypothesis.")
else:
    print(Fore.BLUE + "                    The average wait time in emergency department is 30 minute, fail to reject null hypothesis.")

# Suggestion for: If the wait time significantly exceeds 30 minutes
print(Fore.GREEN + "\n\nIf the wait time significantly exceeds 30 minutes, the hospital should implement measures to reduce waiting times. Here are some changes they could consider:")
print(Fore.BLACK + "           Increase staffing during peak hours to handle more patients efficiently.")
print(Fore.BLACK + "           Implement a triage system to prioritize critical cases and minimize waiting for non-urgent patients.")
print(Fore.BLACK + "           Optimize workflow in the emergency department to improve efficiency.")
print(Fore.BLACK + "           Use technology like patient tracking systems to predict and manage wait times better.")

print(Fore.GREEN + "\n\nFinding is:")
print(Fore.BLUE + "  We could able to see whether we are able to reject or fail to reject the null hypothesis by using different tests like t-test and p-value.")
```

### 6.A survey was conducted on hospital cleanliness and patient satisfaction. The following data was collected:

### Cleanliness  Satisfied  Unsatisfied
### High         90           10
### Medium       60           40
### Low          30           70

### Perform an analysis to check whether hospital cleanliness and patient satisfaction are dependent.

### If cleanliness ratings improve, how do you expect the distribution of satisfied and unsatisfied patients to change

```python
## State the Null and Alternate Hypothesis
print(Fore.GREEN + Style.BRIGHT + "Step 1: State the Hypothesis:")
print(Fore.BLACK + "Null hypothesis is: Hospital cleanliness and patient satisfaction are independent.")
print(Fore.BLACK + "\nAlternate hypothesis is: Hospital cleanliness and patient satisfaction are dependent.")

print(Fore.GREEN + "\n\nStep 2: Create the dataframe from the given data")
data = {'Cleanliness': ['High', 'Medium', 'Low'],
        'Satisfied': [90, 60, 30],
        'Unsatisfied': [10, 40, 70]}
df = pd.DataFrame(data)
print(Fore.BLACK + f"{df}")

## Create the contingency table for cleanliness and satisfaction
## We define the data as a 2D NumPy array, where the rows represent the cleanliness ratings (High, Medium, Low) and the columns represent the number of satisfied and unsatisfied patients.
## Rows: High, Medium, Low cleanliness; Columns: Satisfied, Unsatisfied
print(Fore.GREEN + "\n\nStep 3: Create the contingency table for cleanliness and satisfaction")
contingency_table = df[['Satisfied', 'Unsatisfied']].values
print(Fore.BLACK + f"{contingency_table}")

## This is used for categorical data, where we are examining whether the distribution of categories depends on another categorical variable.
## Perform the Chi-Square Test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)

## Print the results
print(Fore.GREEN + "\n\nStep 4: Print the results of Chi_square and P_Value")
print(Fore.BLUE + "                   Chi-Square Statistic:", chi2_stat)
print(Fore.BLUE + "                   P-value:", p_value)

## Interpretation
print(Fore.GREEN + "\n\nStep 5: The interpretation is:")
if p_value < 0.05:
    print(Fore.BLUE + "                    The cleanliness rating and patient satisfaction are dependent, reject null hypothesis.")
else:
    print(Fore.BLUE + "                    The cleanliness rating and patient satisfaction are independent, fail to reject null hypothesis.")

# Suggestion for: If cleanliness ratings improve, how the distribution of satisfied and unsatisfied patients to change
print(Fore.GREEN + "\n\nIf the cleanliness rating improves, we expect:")
print(Fore.BLACK + "           Higher cleanliness (High): More satisfied patients, fewer unsatisfied patients.")
print(Fore.BLACK + "           Medium cleanliness: A mix, with a larger proportion of satisfied patients than Low.")
print(Fore.BLACK + "           Low cleanliness: Fewer satisfied patients, more unsatisfied.")

print(Fore.GREEN + "\n\nFinding is:")
print(Fore.BLUE + "  We could able to see that by using p-value and significance level we can derive decision like whether to reject or fail to reject null hypothesis.")
```

### 9. The hospital is studying the distribution of patient arrival times in the emergency department. Historical data suggests that emergency cases arrive at an average rate of 5 per hour.

### Model this scenario using an appropriate probability distribution.
### What is the probability that exactly 3 emergency cases will arrive in the next hour?
### If a major accident occurs in the city, how would this affect the probability distribution of emergency arrivals?

```python
print(Fore.BLACK + Style.BRIGHT + "For modeling the number of events occurring within a fixed interval of time, we can use Poisson distribution.")

print(Fore.GREEN + "\n\nPoisson Distribution Formula:")
print(Fore.MAGENTA + "            P(X=k) = (λ^k.e^−λ)/k!")
print(Fore.RED + "\n             P(X=k) is the probability of observing k events")
print(Fore.RED + "             λ is the average number of events in the interval (in your case, 5 events per hour)")
print(Fore.RED + "             e is the base of the natural logarithm (approximately 2.718)")
print(Fore.RED + "             k is the number of events you're interested in (in this case, 3 events)")

print(Fore.GREEN + "\n\nTo model the scenario using poisson distribution, we are taking 10 samples and calculating")
# Given parameters
lambda_rate = 5  # average arrival rate (5 cases per hour)

# Range of number of cases we want to calculate probabilities for (0 to 10)
k_values = np.arange(0, 11)

# Calculate the probabilities for each value of k using the Poisson distribution
probabilities = stats.poisson.pmf(k_values, lambda_rate)

# Display the probabilities for each value of k
for k, prob in zip(k_values, probabilities):
    print(Fore.BLUE + f"    Probability of exactly {k} cases: {prob:.4f}")


print(Fore.GREEN + "\n\nNow if we want to see the probability of exactly 3 cases in next hour we need to calculate that using poisson distribution")
#Given parameters are
lambda_rate = 5  # average arrival rate (5 per hour)
k = 3  # exact number of events (cases)

# Poisson distribution probability for exactly 3 arrivals
probability_3_cases = stats.poisson.pmf(k, lambda_rate)
print(Fore.BLACK + f"    Probability of exactly 3 emergency cases arriving in the next hour: {probability_3_cases:.4f}")

print(Fore.GREEN + "\n\nNow if a major accident occurs in the city, we can see how the probability of exactly 3 cases in next hour is impacting")
# New rate after a major accident (let's assume 10 cases per hour)
lambda_rate_accident = 10

# Poisson distribution probability for exactly 3 arrivals under the new rate
probability_3_cases_accident = stats.poisson.pmf(k, lambda_rate_accident)

print(Fore.BLACK + f"     Probability of exactly 3 emergency cases arriving in the next hour after a major accident: {probability_3_cases_accident:.4f}")

print(Fore.GREEN + "\n\nFinding is:")
print(Fore.BLUE + " So, as a summary if accident cases increase and number of cases also increase per hour , then probability of getting 3 cases in next hour will decrease.")
```

### 10. The number of surgeries performed per day in the hospital follows a specific distribution pattern. Historical data shows the following frequencies:

### Surgeries Performed       Frequency

### 0                          5

### 1                          12

### 2                          18

### 3                          22

### 4                          15

### 5                           8

### Identify and justify the type of probability distribution that best fits this data.
### Calculate the expected number of surgeries performed per day.
### If a new surgical team is hired, how will this affect the probability distribution of daily surgeries?

```python
# Given data
surgeries_performed = np.array([0, 1, 2, 3, 4, 5])
frequencies = np.array([5, 12, 18, 22, 15, 8])

# Total number of days (total frequency count)
total_days = np.sum(frequencies)

# Probability distribution
probabilities = frequencies / total_days

# Output results
print(Fore.GREEN + Style.BRIGHT + "To understand what probability is suits best calculating the probability")
print(Fore.BLACK + f"\n  Probability distribution (Surgeries vs Probability):")
for surgeries, prob in zip(surgeries_performed, probabilities):
    print(Fore.BLUE + f"      {surgeries} surgeries: Probability = {prob:.4f}")

print(Fore.GREEN + "\n\nTo get the expected number of surgeries per day")
# Expected number of surgeries (mean)
expected_surgeries = np.sum(surgeries_performed * probabilities)

print(Fore.BLACK + f"   Expected number of surgeries per day: {expected_surgeries:.2f}")

# Effect of new surgical team
print(Fore.GREEN + "\n\nIf a new surgical team is hired,the impact will be:")
print(Fore.BLACK + "     If the new team increases the number of surgeries, we would expect the mean to increase.")
print(Fore.BLACK + "     This could make the distribution shift to higher surgery counts or increase the variance.")
print(Fore.BLACK + "     The exact effect depends on the new surgery rate, which isn't given here.")

print(Fore.GREEN + "\n\nFinding is:")
print(Fore.BLUE + "  Here, we could able to see which distribution suits better based on probability value.")
```
